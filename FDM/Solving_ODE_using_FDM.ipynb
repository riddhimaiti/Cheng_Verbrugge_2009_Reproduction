```python
import numpy as np
import matplotlib.pyplot as plt
```

We are going to solve the following ODE using Finite Difference Method:

$\frac{d^2 u}{d x^2} = Kx , 0<x<1$

at $x = 0 , u = 0$

at $x = 1 , \frac{d u}{d x} = 1$

```python
def solve_ode(N=200, K=5):
    dx = 1.0/N
    x = np.linspace(0,1,N+1)
    A = np.zeros((N+1,N+1))
    A[0,0]=1 # diagonal
    for i in range(1,N):
        A[i,i-1]=1 # subdiagonal
        A[i,i]=-2 # diagonal
        A[i,i+1]=1 # superdiagonal
    A[N,N-1]=2 # subdiagonal
    A[N,N]=-2 # diagonal
    u = np.zeros(N+1)
    b = np.zeros(N+1)
    for n in range(1,N+1):
        b[n] = K*n*dx**3
    b[N]=K*N*dx**3-2*dx
    # Au=b
    u = np.linalg.solve(A, b)
    return u
```

```python
N=50
K=2
x = np.linspace(0,1,N+1)
u=solve_ode(N,K)
plt.plot(x,u,'r*')
plt.xlabel('x')
plt.ylabel('u')
plt.title('u vs x curve (from FDM)')
```

The Analytical Solution is:

$u(x)=\frac{Kx^3}{6}+(1-\frac{K}{2})x$

```python
U=K*(x**3)/6+(1-K/2)*x
plt.plot(x,U,'b')
plt.xlabel('x')
plt.ylabel('u')
plt.title('u vs x curve (from Analytical)')
```

```python
plt.plot(x,u,'r*',label='FDM')
plt.plot(x,U,'b',label='Analytical')
plt.legend()
plt.xlabel('x')
plt.ylabel('u')
plt.title('u vs x curve')
```

Fully superposing eachother. Hence verified that solution from FDM is same as Analytical solution
